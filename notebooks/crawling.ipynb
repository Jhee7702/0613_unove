```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd

options = webdriver.ChromeOptions()
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

url = "https://www.oliveyoung.co.kr/store/goods/getGoodsDetail.do?goodsNo=A000000223723"
driver.get(url)
time.sleep(3)  # 페이지 로딩 대기

reviews = []
page = 1

while True:
    print(f"크롤링 중: {page} 페이지")
    time.sleep(2)  # 페이지 안정화
    
    # 리뷰 리스트 가져오기
    review_items = driver.find_elements(By.CSS_SELECTOR, '#reviewList li.review-list-item')
    if not review_items:
        print("더 이상 리뷰가 없습니다.")
        break
    
    for item in review_items:
        try:
            content = item.find_element(By.CSS_SELECTOR, 'p.review-cont').text.strip()
            rating = item.find_element(By.CSS_SELECTOR, 'div.review-grade span.grade').text.strip()
            date = item.find_element(By.CSS_SELECTOR, 'div.review-day').text.strip()
            reviews.append({'content': content, 'rating': rating, 'date': date})
        except Exception as e:
            print(f"리뷰 데이터 추출 중 오류: {e}")
            continue

    # 다음 페이지 버튼 찾기
    try:
        next_btn = driver.find_element(By.CSS_SELECTOR, '.paginate > a.btn-next')
        if 'disabled' in next_btn.get_attribute('class'):
            print("마지막 페이지입니다.")
            break
        next_btn.click()
        page += 1
        time.sleep(2)
    except Exception as e:
        print(f"다음 페이지 이동 오류: {e}")
        break

driver.quit()

# 결과 저장
df = pd.DataFrame(reviews)
df.to_csv('onove_perfume_reviews.csv', index=False, encoding='utf-8-sig')
print("크롤링 완료. 파일 저장됨: onove_perfume_reviews.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
import re

options = webdriver.ChromeOptions()
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

url = "https://www.oliveyoung.co.kr/store/goods/getGoodsDetail.do?goodsNo=A000000223723"
driver.get(url)

# 리뷰 탭 클릭
try:
    review_tab = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "li#reviewInfo > a.goods_reputation"))
    )
    review_tab.click()
except Exception as e:
    print("리뷰 탭 클릭 실패:", e)
    driver.quit()
    exit()

time.sleep(3)  # 리뷰 로딩 대기

reviews = []
page = 1

while True:
    print(f"크롤링 중: {page} 페이지")
    
    # 리뷰 아이템 대기
    try:
        review_items = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div.review_cont'))
        )
    except Exception:
        print("리뷰 로딩 실패 혹은 더 이상 리뷰가 없습니다.")
        break

    for item in review_items:
        try:
            rating_text = item.find_element(By.CSS_SELECTOR, '.score_area .point').text.strip()
            # 숫자만 추출 (예: '5점만점에 3점' → 3)
            rating = int(re.search(r'(\d+)점', rating_text).group(1))
            
            date = item.find_element(By.CSS_SELECTOR, '.score_area .date').text.strip()
            option = item.find_element(By.CSS_SELECTOR, 'p.item_option').text.strip()
            content = item.find_element(By.CSS_SELECTOR, '.txt_inner').text.strip()

            reviews.append({
                'rating': rating,
                'date': date,
                'option': option,
                'content': content
            })
        except Exception as e:
            print("리뷰 데이터 추출 오류:", e)
            continue

    # 다음 페이지 버튼 체크 및 클릭
    try:
        next_btn = driver.find_element(By.CSS_SELECTOR, '.paginate a.btn-next')
        if 'disabled' in next_btn.get_attribute('class'):
            print("마지막 페이지입니다.")
            break
        next_btn.click()
        page += 1
        time.sleep(3)  # 페이지 로딩 대기
    except Exception as e:
        print("다음 페이지 이동 오류 또는 마지막 페이지:", e)
        break

driver.quit()

# 결과 저장
df = pd.DataFrame(reviews)
df.to_csv('onove_perfume_reviews.csv', index=False, encoding='utf-8-sig')
print("크롤링 완료. 파일 저장됨: onove_perfume_reviews.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
import re
import os

options = webdriver.ChromeOptions()
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

url = "https://www.oliveyoung.co.kr/store/goods/getGoodsDetail.do?goodsNo=A000000223723"
driver.get(url)
time.sleep(3)

# 상품평 탭 클릭
driver.find_element(By.CSS_SELECTOR, "li#reviewInfo a.goods_reputation").click()
time.sleep(3)

reviews = []
page = 1

while True:
    print(f"크롤링 중: {page} 페이지")
    time.sleep(2)

    review_items = driver.find_elements(By.CSS_SELECTOR, 'div.review_cont')
    if not review_items:
        print("더 이상 리뷰가 없습니다.")
        break

    for item in review_items:
        try:
            rating_text = item.find_element(By.CSS_SELECTOR, '.score_area .point').text.strip()
            rating_match = re.search(r'(\d+)점', rating_text)
            rating = int(rating_match.group(1)) if rating_match else None

            date = item.find_element(By.CSS_SELECTOR, '.score_area .date').text.strip()

            option_elems = item.find_elements(By.CSS_SELECTOR, 'p.item_option')
            option = option_elems[0].text.strip() if option_elems else ''

            content = item.find_element(By.CSS_SELECTOR, '.txt_inner').text.strip()

            reviews.append({
                'rating': rating,
                'date': date,
                'option': option,
                'content': content
            })
        except Exception as e:
            print("리뷰 데이터 추출 오류:", e)
            continue

    try:
        # 현재 페이지 번호 확인
        current_page = int(driver.find_element(By.CSS_SELECTOR, 'div.pageing strong').text)

        # 모든 페이지 링크 가져오기
        page_links = driver.find_elements(By.CSS_SELECTOR, 'div.pageing a[data-page-no]')

        next_page_clicked = False
        for link in page_links:
            if int(link.get_attribute("data-page-no")) == current_page + 1:
                link.click()
                page += 1
                next_page_clicked = True
                time.sleep(2)
                break

        # 다음 페이지로 넘어가는 링크가 없으면 다음 10페이지 버튼 클릭
        if not next_page_clicked:
            next_btns = driver.find_elements(By.CSS_SELECTOR, 'div.pageing a.next')
            if next_btns:
                next_btns[0].click()
                time.sleep(2)
            else:
                print("마지막 페이지입니다.")
                break

    except Exception as e:
        print("페이지 이동 오류:", e)
        break



df = pd.DataFrame(reviews)
df.to_csv('../data/unove_perfume_reviews3.csv', index=False, encoding='utf-8-sig')
print("크롤링 완료. 파일 저장됨: ../data/unove_perfume_reviews3.csv")
```
